## Problem 1: Inside the Models in Scikit-learn

### Part (a): Implement the models by hand
In this part you **may not** use ``sklearn`` for any task.

In the following, you may use ``numpy`` arrays instead of ``pandas`` dataframes.

- Implement a funtion ``split``, which satifies:
    - input: an ``nx2`` dataframe ``data``, a float ``m``
    - return: an ``nx2`` dataframe ``train`` and an ``nx2`` dataframe ``test``, consisting of ``m`` percent and ``100 - m`` percent of the data, respectively.


- Implement K-Nearest Neighbour for predicting a quantitative variable. That is, write a function, ``knn_predict``, that satisfies:
    - input: an integer ``k``, an ``n x 2`` dataframe training set ``train``, an ``n x 1`` dataframe testing set ``test``
    - return: an ``nx2`` dataframe, whose first column is that of ``test`` and whose second column is the predicted values.



- Implement linear regression for predicting a quantitative variable. That is, write a function ``linear_reg_fit`` that satisfies:
    - input: an ``nx2`` dataframe training set ``train``
    - return: the coefficients of the linear regression model - a float ``slope`` and a float ``intercept``.
    
    
- Write a function ``linear_reg_predict`` that satisfies:
    - input: an ``nx1`` dataframe testing set ``test``, as well as the coefficients of the linear regression model
    - return: an ``nx2`` dataframe, whose first column is that of ``test`` and whose second column is the predicted values.
    
    
- Implement a function ``score`` that satisfies:
    - input: an ``nx2`` dataframe ``predicted``, an ``nx2`` dataframe ``actual`` 
    - return: R^2 coefficient of the fit of the predicted values.

**Solution:**

```python
# define a function to split dataframe to train and test consisting of m percent and 100 - m percent respectively.
# test_n2: n by 2 dataframe
# test_n1: n by 1 dataframe with only column "x"

df = pd.read_csv("dataset_1_full.txt")
def split(df, m):
    train = df.sample(frac = m*0.01)
    test_n2 = df.drop(train.index)
    return train, test_n2
train, test_n2 = split(df, 70)
test_n1 = pd.DataFrame(data = test_n2['x'])
```

```python
def knn_predict(k, train, test_n1):
    # Make each columns array-like for comparison later
    x_train = train['x'].values
    y_train = train['y'].values
    x_test = test_n1['x'].values

    # Create an array to store the predicted y values
    y_predict = np.ones(len(test_n1))

    # Compare all the x from test dataset to all x in the train dataset using matrix subtraction
    # Save the euclidean distance in a variable named "distance"
    for i in range(len(test_n1)):
        distance = abs(x_train - x_test[i])
    # sort distance from small to large and store the indices in an array
        index_array = np.argsort(distance)
    # Sort the y from train dataset based on the indices from sorting x
        sorted_y = y_train[index_array]
    # the predicted y value is equal to the average of the k nearest neighbors
        y_predict[i] = sorted_y[:k].mean()
    # Construct a nx2 dataframe
    arrays = {'x': x_test, 'y': y_predict}
    test_knn = pd.DataFrame(data=arrays)
    return test_knn
```

```python
def linear_reg_fit(train):
    # Extract x and y values from corresponding columns.
    x_train = train['x'].values
    y_train = train['y'].values
   
    # Estimate the coefficients
    beta1 = np.sum((x_train - np.mean(x_train)) * (y_train - np.mean(y_train)))/np.sum( (x_train - np.mean(x_train))**2)
    beta0 = np.mean(y_train) - beta1*np.mean(x_train)
    return beta0, beta1

def linear_reg_predict(test_n1, beta0, beta1):
    y_predict = beta0 + beta1 * test_n1
    test_linear = pd.DataFrame({'x': test_n1['x'].values,
                                'y': y_predict['x'].values
                               })
    return test_linear
```

```python
def score(predicted, actural):
    TSS = sum((actural.iloc[:, 1] - actural['y'].mean()) ** 2)
    RSS = sum((actural['y'].values - predicted['y'].values) ** 2)
    r_sqr = 1 - RSS/TSS
    return r_sqr
```

## Problem 2: Handling Missing Data

In this problem, we will be handling the problem of datasets with missing values. Clearly, we cannot simply remove entire rows or columns that contain missing values. In this problem, we explore two different ways to fill in missing values.

The datasets required for this problem is in the ``dataset`` directory. Each file in the ``dataset`` directory contains a one-dimensional data set, with the first column containing the independent variable X, and the second column containing the dependent variable Y.

In this problem, you **may not** use ``sklearn`` or build-in ``pandas`` functions to **directly fill in missing values**. Usage of these libraries/pakcages for related tasks is fine.

### Part (a): Model Based Data Imputation

- Describe in detail how predictive models for data (like KNN and simple linear regression) can be used to fill in missing values in a data set.


- Implement your scheme. That is, write code (preferably a function ``fill`` or two functions ``fill_knn``, ``fill_lin_reg``), which takes an ``n x 2`` dataframe or array with values missing in the 2nd column and fills in these values using KNN and linear regression. 


- You need to, also, write code to evaluate the quality of the values you've filled in.

**Solution:**

We can first separate the dataset to two subsets: training dataset, meaning the rows with both x and y filled in; testing dataset, meaning the rows with only x filled in and missing y. 
Then we can create either KNN or linear regression model using the training dataset.
Now we can predict the missing y values by putting in x values from the testing dataset into either/both models.

```python
def fill_knn(missing_df, full_df, missing_ind, filled_ind, k):
    #preparing data in array form
    x_train = missing_df.loc[filled_ind, 'x'].values.reshape((len(filled_ind), 1))
    y_train = missing_df.loc[filled_ind, 'y'].values
    x_test = missing_df.loc[missing_ind, 'x'].values.reshape((len(missing_ind), 1))
    y_test = full_df.loc[missing_ind, 'y'].values
    
    #fit knn model
    neighbours = KNN(n_neighbors=k)
    neighbours.fit(x_train, y_train)
    
    #predict y-values
    predicted_y = neighbours.predict(x_test)
    
    #score predictions
    r = neighbours.score(x_test, y_test)
    
    #fill in missing y-values
    predicted_df = missing_df.copy()
    predicted_df.loc[missing_ind, 'y'] = pd.Series(predicted_y, index=missing_ind)
    
    return predicted_df, r

#--------  fill_lin_reg
#input: missing_df (dataframe), full_df (dataframe), missing_ind (indices of missing values), 
#       filled_ind (indices of non-missing values), k (integer)
#output: predicted_df (dataframe), r (float)

def fill_lin_reg(missing_df, full_df, missing_ind, filled_ind):
    #preparing data in array form
    x_train = missing_df.loc[filled_ind, 'x'].values.reshape((len(filled_ind), 1))
    y_train = missing_df.loc[filled_ind, 'y'].values
    x_test = missing_df.loc[missing_ind, 'x'].values.reshape((len(missing_ind), 1))
    y_test = full_df.loc[missing_ind, 'y'].values
    
    #fit linear model
    regression = Lin_Reg()
    regression.fit(x_train, y_train)
    
    #predict y-values
    predicted_y = regression.predict(x_test)
    
    #score predictions
    r = regression.score(x_test, y_test)
    
    #fill in missing y-values
    predicted_df = missing_df.copy()
    predicted_df.loc[missing_ind, 'y'] = pd.Series(predicted_y, index=missing_ind)
    
    return predicted_df, r
```

```python
def find_ind(missing_df, full_df):
    test = missing_df[missing_df.isnull().any(axis=1)]
    missing_ind = test.index
    train = missing_df.drop(test.index)
    filled_ind = train.index
    return missing_ind, filled_ind

fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6), (ax7, ax8), (ax9, ax10), (ax11, ax12)) = plt.subplots(6, 2, figsize=(20, 25))
graph = ((ax1, ax2), (ax3, ax4), (ax5, ax6), (ax7, ax8), (ax9, ax10), (ax11, ax12))

for i in range(1,7):
    missing_df = pd.read_csv("dataset_"+str(i)+"_missing.txt")
    full_df = pd.read_csv("dataset_"+str(i)+"_full.txt")
    missing_ind, filled_ind = find_ind(missing_df, full_df)
    predicted_df_knn, r_knn = fill_knn(missing_df, full_df, missing_ind, filled_ind, 2)
    graph[i-1][0].scatter(predicted_df_knn['x'].values, predicted_df_knn['y'].values, label = "predicted_data")
    graph[i-1][0].scatter(full_df['x'].values, full_df['y'].values, c = 'red', label = 'full_data')
    graph[i-1][0].set_title("Dataset_"+str(i)+": KNN")
    graph[i-1][0].set_xlabel("x")
    graph[i-1][0].set_ylabel("Predicted y/Observed y")
    graph[i-1][0].legend(loc='best')
    
    predicted_df_lin, r_lin = fill_lin_reg(missing_df, full_df, missing_ind, filled_ind)
    graph[i-1][1].scatter(predicted_df_lin['x'].values, predicted_df_lin['y'].values, label = "predicted_data")
    graph[i-1][1].scatter(full_df['x'].values, full_df['y'].values, c = 'red', label = 'full_data')
    graph[i-1][1].set_title("Dataset_"+str(i)+": Linear Regression")
    graph[i-1][1].set_xlabel("x")
    graph[i-1][1].set_ylabel("Predicted y/Observed y")
    graph[i-1][1].legend(loc='best')

    print "The R squared of knn and linear regression for dataset_"+str(i)+" are:", r_knn, r_lin
```

Is the product of two normally distributed variables a normally distributed variable? You should fully explain and support your conlusion. 

**Solution:**

```python

x = np.random.normal(0,1,100000)
y = np.random.normal(0,1,100000)
z = x * y
plt.hist(z, bins = 50)
plt.show()
print scipy.stats.normaltest(x)
print scipy.stats.normaltest(y)
print scipy.stats.normaltest(z)

# Both the histogram and normal test reject the hypothesis that z follows normal distribution.
# So the product of two normally distributed variable is not normally distributed.
```
