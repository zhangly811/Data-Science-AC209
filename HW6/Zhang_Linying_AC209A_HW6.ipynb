```python
ratings_df = pd.read_csv('datasets/dataset_4_ratings.txt', delimiter=',')
ratings_df.head()
```

```python
names_df = pd.read_csv('datasets/dataset_4_movie_names.txt', delimiter='|')
names_df.head()
```

```python
#--------  movie_stats
# A function that extracts the number of likes and total number of ratings for a movie
# Input: 
#      movie_name (an optional parameter containing the exact name of the movie)
#      movie_name_contains (an optional parameter containing a portion of the name of the movie)
# Returns: 
#      total_ratings (the total number of ratings for a movie)
#      likes (the total number of likes for a movie)

def movie_stats(movie_name=None, movie_name_contains=None):
    
    #If given an exact movie name:
    if movie_name is not None:        
        #Find the index of the movie, by name, in the "names" dataframe
        movie_index = names_df[names_df['movie_name'] == movie_name].index[0]
        #Get the id for the movie in the "names" dataframe
        movie_id = names_df.loc[movie_index, 'movie_id']
        #Get all ratings for the movie, by id, in the "ratings" dataframe
        ratings_for_movie = ratings_df[ratings_df['movie_id'] == movie_id]
        #Count the total number of ratings
        total_ratings = len(ratings_for_movie)
        #Count the likes (the 1's)
        likes = ratings_for_movie['rating'].sum()
    
    #Otherwise, if given a partial movie name:
    elif movie_name_contains is not None:
        #Find the index of the movie, by name, in the "names" dataframe
        movie_index = names_df[names_df['movie_name'].str.contains(movie_name_contains)].index[0]
        #Get the id for the movie in the "names" dataframe
        movie_id = names_df.loc[movie_index, 'movie_id']
        #Get all ratings for the movie, by id, in the "ratings" dataframe
        ratings_for_movie = ratings_df[ratings_df['movie_id'] == movie_id]
        #Count the total number of ratings
        total_ratings = len(ratings_for_movie)
        #Count the likes (the 1's)
        likes = ratings_for_movie['rating'].sum()
    
    else:
        total_ratings = 0.
        likes = 0.
    
    return float(total_ratings), likes
```

```python
#Make a list of movie names and their ratings info
likability = []

#Iterate through all the movie names
for name in names_df['movie_name'].values:
    #Get ratings info for movie
    total_ratings, likes = movie_stats(movie_name=name)
    #Add movie info to our list
    likability.append((name, likes, total_ratings, likes / total_ratings))

#Sort our list of movie info by like-percentage, in descending order
sorted_likability = sorted(likability, key=lambda t: t[3], reverse=True)  
#Get the movies with top 25 like-percentage
top_25_movies = sorted_likability[:25]

#Print results of ranking
print 'Top 25 Movies'
print '****************************'
for movie, likes, total_ratings, likable in top_25_movies:
    print movie, ':', likable, '({}/{})'.format(likes, total_ratings)
```

**Analysis:** Using the total number of likes doesn't take into account the number of people who dislike the movie. So for example, if a movie is rated by a lot of viewers and most people rated "dislike", it can still have more "likes" than a movie rated by significantly fewer viewers, but all of them rated "like". In this case, we are not confident that the total number of likes reflects the true popularity of the movie. This problem can be solved by using the proportion of people who like a movie, which is the MLE estimate that maximize the likelihood function. 

$$L(\theta)=\prod_{i=1}^{n}\theta^{Y_i}(1-\theta)^{1-Y_i}$$    
$$l(\theta)=log L(\theta)=\sum_{i=1}^{n}\{Y_ilog(\theta)+(1-Y_i)log(1-\theta)\}$$     
$$\frac{d}{d\theta}l(\theta)=\sum_{i=1}^n\{\frac{Y_i}{\theta}-\frac{1-Y_i}{1-\theta}\}$$     
set $\frac{d}{d\theta}l(\theta)=0$, we can get the MLE of $\hat{\theta}$ is $\bar{y}$, which equals $P(\text{label} = \text{like}|\text{movie}) = \theta_{\text{movie}}$ (maximum could be confirmed through second derivative of $l(\theta)$).  

This method also has caveat: if a movie is only rated by a small number of viewers, the like-percentage may not reflect the true popularity of the movie in a large population. This can be interpreted as overfitting the like-probability due to small sample size. To correct for overfitting, we can incorporate some prior beliefs regarding a movie.

In the following, we've provide you a couple of functions for visualize beta priors and approximating their associated posteriors.

```python
#--------  plot_beta_prior
# A function to visualize a beta pdf on a set of axes
# Input: 
#      a (parameter controlling shape of beta prior)
#      b (parameter controlling shape of beta prior)
#      color (color of beta pdf)
#      ax (axes on which to plot pdf)
# Returns: 
#      ax (axes with plot of beta pdf)

def plot_beta_prior(a, b, color, ax):
    
    #Create a beta-distributed random variable with shape a, b
    rv = sp.stats.beta(a, b)
    #Create values from 0 to 1
    x = np.linspace(0, 1, 100)
    #Plot the beta pdf for values from 0 to 1
    ax.plot(x, rv.pdf(x), '-', lw=2, color=color, label='a=' + str(a) + ', b=' + str(b))
    #Set title, legend etc
    ax.set_title('Beta prior with a=' + str(a) + ', b=' + str(b))
    ax.legend(loc='best')
    
    return ax
```

```python
#--------  sample_posterior
# A function that samples points from the posterior over a movie's 
# likability, given a binomial likelihood function and beta prior
# Input: 
#      a (parameter controlling shape of beta prior)
#      b (parameter controlling shape of beta prior)
#      likes (the number of likes in likelihood)
#      ratings (total number of ratings in likelihood)
#      n_samples (number of samples to take from posterior)
# Returns: 
#      post_samples (a array of points from the posterior)

def sample_posterior(a, b, likes, ratings, n_samples):
    #Samples points from a beta distribution 
    #(the posterior of a binomial likelihood and a beta prior is a beta distribution!)
    post_samples = np.random.beta(a + likes, b + ratings - likes, n_samples)
    return post_samples
```

```python
#--------  find_mode
# A function that approximates the mode of a distribution given a sample from the distribution
# Input: 
#      values (samples from the distribution)
#      num_bins (number of bins to use in approximating histogram)
# Returns: 
#      mode (the approximate mode of the distribution)

def find_mode(values, num_bins):
    
    #Make an approximation (histogram) of the distribution using the samples
    bins, edges = np.histogram(values, bins=num_bins)
    #Find the bin in the histogram with the max height
    max_height_index = np.argmax(bins)
    #Find the sample corresponding to the bin with the max height (the mode)
    mode = (edges[max_height_index] + edges[max_height_index + 1]) / 2.
    
    return mode
```

```python
#A list of beta distribution shapes to try out
beta_shapes = [(1, 1), (0.5, 0.5), (2, 2), (4, 2), (2, 4)]
#Length of the list of shapes
n = len(beta_shapes)

#Plot all the beta pdfs in a row
fig, ax = plt.subplots(1, n, figsize=(20, 4))

#Start the index of the current subplot at 0
ax_ind = 0
#Iterate through all the shapes
for a, b in beta_shapes:
    #Plot the beta pdf for a particular shape
    plot_beta_prior(a, b, 'blue', ax[ax_ind])
    #Increment the subplot index
    ax_ind += 1
    
plt.tight_layout()    
plt.show() 
```

**Analysis:** Adding a prior gives us information about what we believe the popularity is going to be for a movie, or movies. This underlying distribution allows the model better predict the true popularity of a movie even under small sample size, given the prior is true. Beta distributions are appropriate priors because it is a continuous probability distributed within interval [0,1], which is the same interval that our outcome fall in. So beta distributions with different shapes can be interpreted as the distribution of people with different opinions regarding a movie.

Adding a prior to our model avoids overfitting, especially when we only have small sample size. 
The prior we tested are:
a=1, b=1: the probability that people like, dislike, or somewhere in between are all equally likely. This prior doesn't tell us anything about the likability of a movie.
a=0.5, b=0.5: people either strongly like the movie or strongly dislike the movie. Not so many people take a neutral standpoint.
a=2, b=2: Most people are neutral about this movie. Very few people have a strong opinion about the movie.
a=4, b=2: more people tend to like the movie than dislike the movie.
a=2, b=4: more people tend to dislike the movie than like the movie.

To help you visually interpret and assess the quality of your classifiers, we are providing you a function to visualize a set of data along with the decision boundaries of a classifier.

```python
#--------  plot_decision_boundary
# A function that visualizes the data and the decision boundaries
# Input: 
#      x (predictors)
#      y (labels)
#      poly_flag (a boolean parameter, fits quadratic model if true, otherwise linear)
#      title (title for plot)
#      ax (a set of axes to plot on)
# Returns: 
#      ax (axes with data and decision boundaries)

def plot_decision_boundary(x, y, model, poly_flag, title, ax):
    # Plot data
    ax.scatter(x[y == 2, 0], x[y == 2, 1], c='g') # Add another class
    ax.scatter(x[y == 1, 0], x[y == 1, 1], c='b')
    ax.scatter(x[y == 0, 0], x[y == 0, 1], c='r')
    
    # Create mesh
    interval = np.arange(0,1,0.01)
    n = np.size(interval)     #this has to be np.size
    x1, x2 = np.meshgrid(interval, interval)   #this has to be np.meshgrid
    x1 = x1.reshape(-1, 1)
    x2 = x2.reshape(-1, 1)
    xx = np.concatenate((x1, x2), axis=1)

    # Predict on mesh points
    if(poly_flag):
        quad_features = preprocessing.PolynomialFeatures(degree=2)
        xx = quad_features.fit_transform(xx)
    yy = model.predict(xx)    
    yy = yy.reshape((n, n))

    # Plot decision surface
    x1 = x1.reshape(n, n)
    x2 = x2.reshape(n, n)
    ax.contourf(x1, x2, yy, alpha=0.1)
    
    # Label axes, set title
    ax.set_title(title)
    ax.set_xlabel('Latitude')
    ax.set_ylabel('Longitude')
    
    return ax
```

```python
from sklearn.linear_model import LogisticRegression as LogReg

# Fit logistic regression
reg = LogReg(solver = "newton-cg", multi_class="multinomial")
model_log = reg.fit (X, y)
score_log = reg.score(X, y)

# Fit LDA
LDA = da.LinearDiscriminantAnalysis()
model_LDA = LDA.fit(X, y)
score_LDA = LDA.score(X, y)


# Plot these two models on top of a scatter plot
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

plot_decision_boundary(X, y, model_log, False, "Logistic Regression of Year 2000", ax[0])
plot_decision_boundary(X, y, model_LDA, False, "LDA of Year 2000", ax[1])
plt.show()

print "The accuracy of Logistic regression: ", score_log
print "The accuracy of LDA: ", score_LDA
print "Based on accuracy of the two tests, LDA is slightly better than multinomial logistic regression. Since the difference is not significant, we will test both models. "
```

**Answers:**
When p>>n, the performance of LDA is far from ideal, as seen in microarray analyses where
the the numbe rof covariates p is way larger than sample size n. There are two major concerns here. First, the sample covariance matrix estimate is singular and cannot be inverted. Although we may use the generalized inverse instead, the estimate will be very unstable due to lack of observations. Second, high dimensionality makes direct matrix operation formidable, hence hindering the applicability of this method. Therefore, we can improve the model by regularizing the covariance matrix as follows:
$$\sum = \alpha\sum + (1-\alpha)I_p$$
    for some $\alpha$, $0\leq\alpha\leq1$

The difference between LDA and QDA is that: LDA assumes that the covariances of the MVN distributions within classes are equal, and QDA allows them to be different. To find an intermediate fit between LDA and QDA, when can take an average of the covariance matrix of each class, or we can still use the formula above to regularize each covariance matrix so to reduce the flexibility of covariance matrix.
